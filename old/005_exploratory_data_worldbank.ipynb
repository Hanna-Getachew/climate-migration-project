```python
import pandas as pd
```

```python
worldbank_data_somalia = pd.read_csv('data/API_SOM_DS2_en_csv_v2_3285.csv', skiprows=4)
```

```python
worldbank_data_somalia['Indicator Name'].unique()
```

```python
worldbank_data_somalia = worldbank_data_somalia[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2023']]
```

```python
worldbank_data_somalia.dropna(subset=['2023'], how='all', inplace=True)
```

```python
# List of indicators to filter
indicators_to_keep = [
    "Urban population (% of total population)",
    "Rural population growth (annual %)",
    "Food imports (% of merchandise imports)",
    "Mortality rate, infant (per 1,000 live births)",
    "Death rate, crude (per 1,000 people)",
    "Population, total",
    "Droughts, floods, extreme temperatures",
    "Average precipitation in depth (mm per year)",
    "Disaster risk reduction progress score (1-5 scale; 5=best)",
    "CPIA policy and institutions for environmental sustainability rating (1=low to 6=high)"
]

# Filter the dataframe
df_filtered = worldbank_data_somalia[worldbank_data_somalia["Indicator Name"].isin(indicators_to_keep)]
```
