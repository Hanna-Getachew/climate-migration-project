```python
import pandas as pd
```

```python
climate_catastrophes_data = pd.read_excel("data/climate_catastrophes/emdat_catastrophes.xlsx", converters={
    "Start Month": int,
    "End Month": int,
    "Start Day": int,
    "End Day": int,
    "Total Deaths": int,
    "Total Affected": int
})
```

```python
data_inflation = pd.read_csv("data/economic/inflation_somalia_cleaned.csv") 
```

```python
climate_catastrophes_data_subset = climate_catastrophes_data[['Country', 'Start Year', 'Start Month', 'End Year', 'End Month',  'Disaster Type', 'Total Affected']]
```

```python
climate_catastrophes_data_subset["End Year"] = climate_catastrophes_data_subset["End Year"].fillna(climate_catastrophes_data_subset["Start Year"])
climate_catastrophes_data_subset["End Month"] = climate_catastrophes_data_subset["End Month"].fillna(climate_catastrophes_data_subset["Start Month"])
```

```python
climate_catastrophes_data_subset
```

```python
expanded_rows = []
for _, row in climate_catastrophes_data_subset.iterrows():
    for year in range(row["Start Year"], row["End Year"] + 1):  # Iterate through years
        start_month = row["Start Month"] if year == row["Start Year"] else 1
        end_month = row["End Month"] if year == row["End Year"] else 12
        
        for m in range(start_month, end_month + 1):
            new_row = row.copy()
            new_row["month"] = m
            new_row["year"] = year  # Assign the correct year for each row
            expanded_rows.append(new_row)
```

```python
climate_catastrophes_data = pd.DataFrame(expanded_rows)
```

```python
#read all the files in folder data/migration, and concatenate them into one dataframe

import os
import glob

path = 'data/migration'
all_files = glob.glob(os.path.join(path, "*.xlsx"))
df_from_each_file = (pd.read_excel(f) for f in all_files)
migration_data = pd.concat(df_from_each_file, ignore_index=True)
```

```python
migration_data['date'] = pd.to_datetime(migration_data['Date of Assessment'])
migration_data['month'] = migration_data['date'].dt.month
migration_data['year'] = migration_data['date'].dt.year
```

```python
migration_data_grouped = migration_data[['year','month','Total new arrivals since last week']].groupby(['year','month']).sum().reset_index()
```

```python
df = pd.merge(migration_data_grouped, climate_catastrophes_data, on=['year','month'], how='left')
```

```python
df = pd.merge(df, data_inflation, on=['year','month'], how = 'left')
```
