```python
import pandas as pd
```

```python
climate_catastrophes_data = pd.read_excel("data/climate_catastrophes/public_emdat_custom_request_2025-03-26_aed911cc-bd9c-44b0-b429-811b8aff6b88.xlsx", converters={
    "Start Month": int,
    "End Month": int,
    "Start Day": int,
    "End Day": int,
    "Total Deaths": int,
    "Total Affected": int
})
```

```python
climate_catastrophes_data
```

```python
climate_catastrophes_data_subset = climate_catastrophes_data[['Country', 'Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day', 'Disaster Type', 'Total Deaths', 'Total Affected']]
```

```python
climate_catastrophes_data_subset
```

```python
expanded_rows = []
for _, row in climate_catastrophes_data_subset.iterrows():
    for year in range(row["Start Year"], row["End Year"] + 1):  # Iterate through years
        start_month = row["Start Month"] if year == row["Start Year"] else 1
        end_month = row["End Month"] if year == row["End Year"] else 12
        
        for m in range(start_month, end_month + 1):
            new_row = row.copy()
            new_row["month"] = m
            new_row["year"] = year  # Assign the correct year for each row
            expanded_rows.append(new_row)
```

```python
climate_catastrophes_data = pd.DataFrame(expanded_rows)
```

```python
somalia_mobility = pd.read_excel('data/somalia_mobility_iom_2024.xlsx')
somalia_mobility['date'] = pd.to_datetime(somalia_mobility['Date of Assessment'])
somalia_mobility['month'] = somalia_mobility['date'].dt.month
somalia_mobility_grouped = somalia_mobility[['month','Total new arrivals since last week']].groupby('month').sum().reset_index()
somalia_mobility_grouped['year'] = 2024
```

```python
somalia_mobility_grouped
```

```python
data = pd.merge(somalia_mobility_grouped, climate_catastrophes_data, on=['year','month'], how='left')
```
